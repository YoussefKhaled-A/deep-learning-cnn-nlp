## 6) Loops (while) + input validation

```python
# Keep asking until the user enters a positive number
num = int(input("Enter a positive number: "))

while num <= 0:
    print("Not positive. Try again.")
    num = int(input("Enter a positive number: "))

print("Good! You entered:", num)
```

## 10) Libraries + loops + data processing (random)

```python
import random

# Simulate 10 dice rolls and compute average
rolls = []
for _ in range(10):
    rolls.append(random.randint(1, 6))

average = sum(rolls) / len(rolls)

print("Rolls:", rolls)
print("Average:", average)
print("Max roll:", max(rolls))
print("Min roll:", min(rolls))
```
